```python
path_to_project = "/home/dev/projects/paper/"
path_to_load_data = path_to_project + "data/raw/LoadProfile/"
path_to_raw_data = path_to_project + "data/raw/"
```

```python
import pandas as pd
import os
import yaml
```

```python
df_1 = pd.read_csv(path_to_load_data + "LoadProfile_20IPs_2016.csv", header=1, sep=";")
df_1
```

```python
df_1.rename(columns={"Time stamp":"date"}, inplace=True)
df_1.rename(columns={f'LG {str(i)}': f'LP16_{i}' for i in range(1, 21)}, inplace=True)

df_1.date = pd.to_datetime(df_1.date, format="%d.%m.%Y %H:%M:%S", errors='coerce')
df_1[(df_1["date"] >= "2016-03-27 02:00") &(df_1["date"] < "2016-03-27 03:00")] = None
df_1['date'] = df_1['date'].dt.tz_localize('Europe/Berlin', ambiguous=True)
df_1.dropna(subset="date", inplace=True)
df_1['date'] = df_1['date'].dt.tz_convert('UTC')
df_1['date'] = df_1['date'].dt.strftime('%Y-%m-%d %H:%M')
df_1['date'] = pd.to_datetime(df_1.date).dt.tz_localize('UTC')
```

```python
df_1['date'][0]
```

```python
datasets = {}
for col in df_1.columns[1:]:  
    series = df_1[['date', col]]
    series["signal_id"] = "load"
    series.set_index("signal_id", inplace=True)
    series.rename(columns={series.columns[1]:"power"}, inplace=True) 
    datasets[col] = series
    
```

```python
df_2 = pd.read_csv(path_to_load_data + "LoadProfile_30IPs_2017.csv", header=1, sep=";")
df_2
```

```python
df_2.rename(columns={"Time stamp":"date"}, inplace=True)
df_2.rename(columns={f'LG {str(i).zfill(2)}': f'LP17_{i}' for i in range(1, 31)}, inplace=True)
df_2.date = pd.to_datetime(df_2.date, format="%d.%m.%Y %H:%M:%S", errors='coerce')
df_2[(df_2["date"] >= "2017-03-26 02:00:00") &(df_2["date"] < "2017-03-26 03:00:00")] = None
df_2['date'] = df_2['date'].dt.tz_localize('Europe/Berlin', ambiguous=True)
df_2.dropna(subset="date", inplace=True)
df_2['date'] = df_2['date'].dt.tz_convert('UTC')
df_2['date'] = df_2['date'].dt.strftime('%Y-%m-%d %H:%M')
df_2['date'] = pd.to_datetime(df_2.date).dt.tz_localize('UTC')
```

```python
for col in df_2.columns[1:]:  
    series = df_2[['date', col]]
    series["signal_id"] = "load"
    series.set_index("signal_id", inplace=True)
    series.rename(columns={series.columns[1]:"power"}, inplace=True) 
    datasets[col] = series
```

```python
# save datasets in separate parquet files
for ds in datasets:
    df = datasets[ds]
    df = df.dropna()
    file_path = f"{path_to_raw_data}{ds}/electricity_demand.parquet"
    os.makedirs(os.path.dirname(file_path), exist_ok=True)
    df.to_parquet(file_path)
```

```python
# create config files  
for ds in datasets:
    df = datasets[ds] 
    df = df.dropna()
    config_content = {
        '_target_': 'eile_forecast.forecast.split_data.GluontsDatasetEile',
        'name': ds,
        'add_seasonalities': True,
        'add_holidays_and_weekends': True,
        'add_weather': False,
        'benchmark_end_date': str(df.date.max().strftime('%Y-%m-%d %H:%M')),
        'main_signal_name': 'load',
        'forecast_signal': 'load',
        'benchmark_start_date': str(df.date.min().strftime('%Y-%m-%d %H:%M')),
        'country': 'Germany',
    }
    
    # Write to YAML file
    yaml_file_path = f'{path_to_project}conf/datasets/{ds}.yaml'
    with open(yaml_file_path, 'w') as yaml_file:
        yaml.dump(config_content, yaml_file, default_flow_style=False)
```
