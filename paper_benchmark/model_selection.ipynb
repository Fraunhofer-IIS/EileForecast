In EileForecast, we benchmark the most commonly offered forecasting methods in open-source EMS and dedicated STLF software.
Here, we create the statistics to find these out, given our lit review.

```python
# The frequency is the number of occurencies of a model in different software packages aka in how many packages is it offered?
frequency = {"Seasonal Naive": 4, "Naive":1, "Seasonal Average": 2, "ARIMA":2, "Sarima":1, "Sarimax":2 ,"LSTM": 4, "Moving Average":3 , "Linear Regression":3 , "Lasso Regression":1 , "Ridge Regression":1, "Elastic Net":1, "K-Nearest Neighbours":1, "Random Forest Regression":1, "Gradient Boosting Regression":2, "AdaBoost Regression":1, "XGBoost":1, "LightGBM":2, "Quantile Regression":1, "EnergyPatchTST":1, "PatchTST":1, "Autoformer":1, "Decomposition Linear":1, "Informer":1, "Transformer":1, "Prophet":1, "Feed-Forward Neural Network":1, "Simple Exponential Smoothing":1, "Holt Winters": 1, "RNN":1, "GRU":1  }
```

```python
# We cluster the models into groups, and count the number of packages that offer the model group
freq_clusters = { "Transformer":1,"RNN":4, "Quantile Regression":1, "Linear Regression":3, "Gradient Boosting Regression":4, "FFNN":1,  "Distance Based Naive":5,"Decomposition":3, "Average":5, "ARIMA":4, }
```

```python
import matplotlib.pyplot as plt
```

```python
methods = list(freq_clusters.keys())
counts = list(freq_clusters.values())
percentage = [(count / 9) for count in counts]
colors = ['skyblue' if p > 0.3 else 'lightgrey' for p in percentage]
plt.figure(figsize=(20, 10))
plt.barh(methods, percentage, color=colors)
plt.xlabel('Frequency',  fontsize=26)
plt.grid(axis='x')
plt.xticks(fontsize=26)                 
plt.yticks(fontsize=26)
plt.axvline(x=0.3, color='black', linestyle='--', linewidth=3)

plt.savefig("/home/dev/projects/paper/data/outputs/benchmark/results/model_frequencies.pdf", bbox_inches='tight')
plt.show()
```
